```python
data = np.c_[x, y]
```

```python
column = ['sepal_Length', 'sepal_Width', 'petal_Length', 'petal_Width', 'class']
df=pd.DataFrame(data, columns=column)
df.head(4)
```

```python
df['sepal_Length'].agg(['mean','max','count','median','std','var'])
```

```python
replacements ={0:'setosa',1:'vesicolor',2:'virginica'}
```

```python
df['class_rep'] = df['class'].map(replacements)
```
